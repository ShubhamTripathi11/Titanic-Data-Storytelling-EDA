```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("Titanic_Dataset.csv",encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['PassengerId','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
pd.isnull(df).sum()
```

# Exploratory Data Analysis(EDA)

# Survived

```python
sns.countplot(x='Survived',data=df)
plt.title('Survival Distribution')
plt.show()
```

*From above graph we can see that people survived in this incident are very less, where as 0 represent non-survival of people and 1 respresent survival*

# Survival by Class (Pclass)

```python
sns.countplot(x='Pclass',hue='Survived',data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

*From above graoh we can say that upper classes pasangers have higher survival rates*

# Survival by Gender

```python
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title('Survival by Gender')
plt.show()
```

*From above graph we can say that women had a higher survival rate than men*

# Survival by Embarked

```python
sns.countplot(x='Embarked',hue='Survived',data=df)
plt.title('Survival by Embarkation Point')
plt.show()
```

*This plot shows whether the port of embarkation influenced survival chances*

# Age Distribution

```python
sns.histplot(df['Age'],bins=30,kde=True,color='purple')
plt.title('Age Distribution')
plt.show()
```

*From above graph we can see that most of the passenger are of age group 30*

```python
plt.figure(figsize=(20,6))
sns.histplot(df[df['Survived']==1]['Age'],bins=30,color='green',label='Survived',kde=True)
sns.histplot(df[df['Survived']==0]['Age'],bins=30,color='red',label='Not Survived',kde=True)
plt.legend()
plt.title('Survival Distribution by Age')
plt.show()
```

*From above plot we can see that younger passengers have higher survival rates*

# Conclusion:

*1)Class and gender played key roles in Survival.
2)Younger passengers particularly children, had higher survival rates.
3)Embarkation point, passengers who embarked at certain ports had slightly different survival probabilities*
